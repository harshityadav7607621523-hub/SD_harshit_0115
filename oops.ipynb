```python
class Student:
  # attribute data
    def __init__(self, id, name, marks):
        self.id = id
        self.name = name
        self.marks = marks
# methods(functions)
    def display(self):
        print("ID:", self.id)
        print("NAME:", self.name)
        print("MARKS:", self.marks)

#object creation
s1 = Student(101, "Harshit", 98)
s1.display()
```

```python
#inheritance
class parent:
  def test_parent(self):
    print("this is my parent class")

```

```python
class child(parent):
  pass
```

```python
child_obj = child()
```

```python
child_obj.test_parent()
```

```python
# Parent class
class Class1:
    def test_class1(self):
        print("This is Class1 method.")


# Child class
class Class2(Class1):
    def test_class2(self):
        print("This is Class2 method.")


# Another child class
class Class3(Class2):
    def test_class3(self):
        print("This is Class3 method.")


# Create object of Class3
obj_class3 = Class3()

# Call methods
obj_class3.test_class1()   # Inherited from Class1
obj_class3.test_class2()   # Inherited from Class2
obj_class3.test_class3()   # Class3's own method
```
