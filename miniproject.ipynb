```python
def calc(a, b, op):
    if op == "+":
        print("Answer =", a + b)

    elif op == "-":
        print("Answer =", a - b)

    elif op == "*":
        print("Answer =", a * b)

    elif op == "/":
        if b != 0:
            print("Answer =", a / b)
        else:
            print("0 se divide nahi kar sakte.")

    else:
        print("Invalid Operator")


# Main Program
n1 = int(input("Enter First Number: "))
n2 = int(input("Enter Second Number: "))
op = input("Enter Operator (+, -, *, /): ")

calc(n1, n2, op)
```
