```python
import time
from IPython.display import clear_output
from zoneinfo import ZoneInfo
from datetime import datetime


while True:
    clear_output(wait=True)
    its_time=datetime.now(ZoneInfo("Asia/Kolkata"))
    print("Date:",its_time.strftime("%d-%m-%y"))
    print("Day :",its_time.strftime("%A"))
    print("Time:",its_time.strftime("%I:%M:%S %p"))
    print("zone:","India standard time(IST)")
    # current_time = datetime.now().strftime("%I:%M:%S %p")
    # zoneinfo=datetime.now().strftime("Asia/Kolkata")


    # print("     DIGITAL CLOCK")
    # print(current_time)
    time.sleep(1)
```
